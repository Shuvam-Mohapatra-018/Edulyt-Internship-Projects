```python
import seaborn as sns
df_new=df.drop(['first_name', 'last_name', 'email','gender','address','INT_ID','Prev_ID','AGT_ID'], axis=1)
df_new['Net_income']=df_new['Salary']-df_new['ExDebt (Liability)']
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
df_new['empstaus']= le.fit_transform(df_new['empstaus']) #SE=1,NSE=0
df_new['ExCus (Customer in Past)']= le.fit_transform(df_new['ExCus (Customer in Past)']) #Y=1,N=0
df_new['tdecision']= le.fit_transform(df_new['tdecision'])#Approved=0,Decline=1,pending=2


df_new = df_new[df_new.tdecision != 2]
df_new.reset_index(inplace = True, drop = True)

df_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X = df_new[[ 'Net_income','age','ExCus (Customer in Past)','empstaus']]
y = df_new['tdecision']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,test_size=0.15)

clf = DecisionTreeClassifier().fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', C=1000,degree=1,gamma=0.01).fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
min_max_scaler = preprocessing.MinMaxScaler()
def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(min_max_scaler.fit_transform(pd.DataFrame(df_new[col])),columns=[col])
    return df

scaled_df = scaleColumns(df_new,['age','Net_income','Salary','Net_income','Booking_Amt'])
scaled_df.head()
```

```python
scaled_df.corr()
```

From COORELATION MATRIX WE CAN OBSERVE THAT tdecision depends on Net Income/Salary and Age

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.calibration import CalibratedClassifierCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
X1 = scaled_df[[ 'Net_income','age','ExCus (Customer in Past)','empstaus']]
y1 = scaled_df['tdecision']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, random_state=0,test_size=0.15)

calibrated_forest = CalibratedClassifierCV(base_estimator=RandomForestClassifier(n_estimators=10))
pipe = Pipeline([('select', SelectKBest()),('model', calibrated_forest)])
param_grid = {'select__k': [1, 2,3,4],'model__base_estimator__max_depth': [2,3, 4,5, 6,7, 8]}
search = GridSearchCV(pipe, param_grid, cv=10).fit(X1, y1)
search.cv_results_
print(search.best_params_)
print(search.best_estimator_)
```
